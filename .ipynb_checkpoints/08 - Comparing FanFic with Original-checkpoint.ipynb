```python
from bookworm import *
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12,9)

import pandas as pd
import numpy as np
```

```python
original = load_book('data/raw/hp1.txt')
characters_original = extract_character_names(original)
sequences_original = get_sentence_sequences(original)

df_original = find_connections(sequences_original, characters_original)
cooccurence_original = calculate_cooccurence(df_original)
```

```python
import networkx as nx
interaction_df_original = get_interaction_df(cooccurence_original, threshold=5)
print(interaction_df_original.sample(5))
G_original = nx.from_pandas_edgelist(interaction_df_original,source='source',target='target')
```

```python
nx.draw_spring(G_original,node_size=7)
```
